## 1. Environment Setup & Dependency Loading

Let's import pandas, matplotlib, seaborn, and our internal configuration/preprocessing modules.

```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Add src to Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..", "src")))
import config
from preprocess import get_preprocessing_steps, preprocess_text
```

## 2. Dataset Loading & Inspection

We load the generated dataset containing 7,500 rows and look at its columns and label distributions.

```python
# Load dataset
df = pd.read_csv(config.DATASET_CSV)
print(f"Dataset columns: {df.columns.tolist()}")
print(f"Total rows: {len(df)}")
df.head()
```

Let's visualize the class distribution to ensure perfect balance across all 5 classes.

```python
plt.figure(figsize=(8, 5))
colors = [config.CATEGORY_COLORS[cat] for cat in config.CATEGORIES]
sns.countplot(data=df, x='label', order=config.CATEGORIES, palette=colors)
plt.title('Article Category Distributions')
plt.xlabel('Category')
plt.ylabel('Article Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 3. Text Preprocessing Pipeline Walkthrough

Let's trace a sample news article text through our 7-step NLTK preprocessing pipeline to verify its transformations.

```python
sample_text = "The National Assembly passed the 2025 Appropriation Bill. Visit https://tasued.edu.ng for news!"
steps = get_preprocessing_steps(sample_text)

print("Preprocessing Pipeline Steps Trace:\n")
for step_name, val in steps.items():
    print(f"[{step_name.upper()}]:")
    print(f"  {val}\n")
```

## 4. Model Performance Comparison

We load `results/evaluation_results.csv` and plot the accuracy comparison across our four models.

```python
df_results = pd.read_csv(config.EVAL_RESULTS_CSV)
df_results
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_results, x='Classifier', y='Accuracy', palette='Blues_r')
plt.title('Classifier Accuracy Comparison (Stratified Test Set)')
plt.ylabel('Test Accuracy (Fraction)')
plt.ylim(0.75, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for index, row in df_results.iterrows():
    plt.text(index, row['Accuracy'] + 0.005, f"{row['Accuracy']*100:.2f}%", ha='center', fontweight='bold')
plt.show()
```

## 5. Confusion Matrix Visualization

Let's load the confusion matrix for the best-performing model (SVM) and plot it as a heatmap.

```python
df_cm = pd.read_csv(config.CONFUSION_MATRIX_CSV, index_col=0)
plt.figure(figsize=(8, 6))
sns.heatmap(df_cm, annot=True, fmt='d', cmap='Purples', cbar=True)
plt.title('Confusion Matrix Heatmap — Calibrated SVM')
plt.ylabel('Actual Category')
plt.xlabel('Predicted Category')
plt.show()
```

### Data Analysis Key Findings
- The classification dataset is perfectly balanced with exactly 1,500 news articles per category (Politics, Sports, Technology, Entertainment, and Business), making up 7,500 rows in total.
